````python
# Thêm các thư viện cần thiết
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import requests as rq
from datetime import datetime, timedelta
from bs4 import BeautifulSoup
import csv
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from pmdarima import auto_arima
from sklearn.metrics import mean_absolute_error, mean_squared_error
````

````python
# Đặt ngày bắt đầu và ngày kết thúc
start_date = datetime(2020, 1, 1)
end_date = datetime(2024, 10, 15)

# Tạo khoảng thời gian ngày
date_range = pd.date_range(start=start_date, end=end_date)

# Đường dẫn cơ bản để lấy dữ liệu cổ phiếu
base_url = 'https://msh-appdata.cafef.vn/rest-api/api/v1/MatchPrice?symbol=VIC&date='

# Danh sách để lưu trữ dữ liệu cổ phiếu
stock_data = []

# Lặp qua từng ngày trong khoảng thời gian
for current_day in date_range:
    date_str = current_day.strftime('%Y%m%d')  # Chuyển đổi ngày sang định dạng chuỗi YYYYMMDD
    full_url = base_url + date_str

    try:
        response = rq.get(full_url)
        response.raise_for_status()  # Nâng cao lỗi cho các phản hồi không hợp lệ

        # Trích xuất dữ liệu từ phản hồi
        stock_info = response.json().get('data')
        if stock_info:
            # Tạo DataFrame từ dữ liệu cổ phiếu
            df_stock = pd.DataFrame(stock_info)
            df_stock = df_stock.sort_values('tradeDate')  # Sắp xếp dữ liệu theo ngày giao dịch

            # Thu thập dữ liệu cổ phiếu liên quan
            stock_data.append({
                'tradeDate': current_day.strftime('%Y-%m-%d'),  # Ngày giao dịch
                'Open': df_stock.loc[df_stock['price'] > 0, 'price'].iloc[0],  # Giá mở cửa
                'Low': df_stock['price'].min(),  # Giá thấp nhất
                'High': df_stock['price'].max(),  # Giá cao nhất
                'Close': df_stock['price'].iloc[-1],  # Giá đóng cửa
                'Volume': df_stock['totalVolume'].max()  # Khối lượng giao dịch
            })
    except rq.exceptions.RequestException as e:
        print(f"Lỗi khi lấy dữ liệu cho {current_day}: {e}")

# Chuyển danh sách dữ liệu cổ phiếu thành DataFrame và lưu vào file CSV
df_VIC = pd.DataFrame(stock_data)
df_VIC.to_csv('data_VIC_ADY.csv', index=False)
````

````python
# Hàm để lấy dữ liệu từ từng trang
def get_events_data(page_index):
    url = f"https://s.cafef.vn/Ajax/Events_RelatedNews_New.aspx?symbol=VIC&floorID=0&configID=0&PageIndex={page_index}&PageSize=30&Type=2"
    response = rq.get(url)
    
    # Kiểm tra xem phản hồi có thành công không
    if response.status_code != 200:
        print(f"Failed to retrieve data for page {page_index}")
        return None
    
    return response.content

# Hàm để phân tích HTML và trích xuất sự kiện
def parse_events_data(html_content):
    soup = BeautifulSoup(html_content, 'html.parser')
    events = []
    
    # Tìm tất cả các thẻ sự kiện
    event_list = soup.find_all('li')  # Giả sử mỗi sự kiện nằm trong thẻ <li>
    
    for event in event_list:
        # Lấy ngày giờ sự kiện (nằm trong thẻ <span>)
        datetime_str = event.find('span').text.strip()
        
        # Lấy tiêu đề sự kiện
        title = event.text.strip().replace(datetime_str, "").strip()
        
        # Chuyển đổi thành kiểu datetime và chỉ lấy phần ngày
        date = pd.to_datetime(datetime_str, dayfirst=True).date()  # Giữ chỉ phần ngày
        
        events.append([date, title])  # Thay đổi datetime thành date

    return events

# Hàm chính để lặp qua các trang và lưu dữ liệu
def scrape_events_to_csv(output_file):
    all_events = []
    page_index = 1

    while True:  
        html_content = get_events_data(page_index)
        
        if html_content:
            events = parse_events_data(html_content)
            if not events:  # Nếu không có sự kiện nào, dừng lại
                break
            all_events.extend(events)
            page_index += 1  # Tăng chỉ số trang
        else:
            break
    
    # Lưu dữ liệu vào file CSV với encoding UTF-8
    with open(output_file, mode='w', newline='', encoding='utf-8-sig') as file:
        writer = csv.writer(file)
        writer.writerow(['date', 'title'])  
        writer.writerows(all_events)

    print(f"Data has been successfully saved to {output_file}")

scrape_events_to_csv('events_data.csv')
````

````python
# Thiết lập thông tin cần thiết
symbol = "VNINDEX"
start_date = "01/01/2023"
end_date = "12/31/2023"
page_size = 20
total_pages = 14  # Số trang tối đa mà bạn biết

# Khởi tạo danh sách để lưu trữ dữ liệu
all_data = []

for page_index in range(1, total_pages + 1):
    # Tạo URL cho mỗi trang
    url = f"https://s.cafef.vn/Ajax/PageNew/DataHistory/PriceHistory.ashx?Symbol={symbol}&StartDate={start_date}&EndDate={end_date}&PageIndex={page_index}&PageSize={page_size}"
    
    # Gửi yêu cầu HTTP
    response = rq.get(url)

    # Kiểm tra xem yêu cầu có thành công không
    if response.status_code != 200:
        print(f"Lỗi khi lấy dữ liệu từ trang {page_index}: {response.status_code}")
        continue  # Bỏ qua trang này nếu có lỗi

    data = response.json()  # Chuyển đổi phản hồi thành định dạng JSON

    # Kiểm tra xem có dữ liệu hay không
    if 'Data' in data and 'Data' in data['Data']:
        all_data.extend(data['Data']['Data'])  # Lưu dữ liệu vào danh sách
    else:
        print(f"Không có dữ liệu từ trang {page_index}.")
    

# Chuyển đổi dữ liệu thành DataFrame
df_vnindex = pd.DataFrame(all_data)

# Lưu dữ liệu vào file CSV
df_vnindex.to_csv('vn_index_data_2023.csv', index=False)
print("Dữ liệu đã được lưu vào vn_index_data_2023.csv")
````

- **Xử lý outliers:** Tính toán các giá trị tứ phân vị (Q1, Q3) và khoảng tứ phân vị (IQR) để xác định giới hạn dưới và trên cho từng cột giá và khối lượng. Những dòng dữ liệu có giá trị nằm ngoài giới hạn này đã được loại bỏ nhằm cải thiện chất lượng dữ liệu.

- **Tính toán các chỉ số mới:** Phần trăm thay đổi giữa giá mở cửa và giá đóng cửa (Daily Change (%)) cũng như phần trăm thay đổi giữa các ngày liên tiếp (Percent Change) đã được thêm vào để cung cấp thêm thông tin về xu hướng giá.

- **Lưu dữ liệu đã làm sạch:** Cuối cùng, dữ liệu đã được làm sạch và xử lý được lưu vào một file CSV mới để thuận tiện cho các bước phân tích tiếp theo.


````python
# Đọc dữ liệu từ file CSV
df_VIC = pd.read_csv('data_VIC_ADY.csv')

# Chuyển đổi cột tradeDate thành kiểu datetime
df_VIC['tradeDate'] = pd.to_datetime(df_VIC['tradeDate'])

# Xóa các dòng có giá trị bị thiếu
df_VIC.dropna(inplace=True)

# Tính các giá trị tứ phân vị đầu tiên (Q1) và thứ ba (Q3)
Q1 = df_VIC[['Open', 'Low', 'High', 'Close', 'Volume']].quantile(0.25)
Q3 = df_VIC[['Open', 'Low', 'High', 'Close', 'Volume']].quantile(0.75)
IQR = Q3 - Q1  # Khoảng tứ phân vị (Interquartile Range)

# Xác định giới hạn dưới và trên để loại bỏ outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Lọc dữ liệu, loại bỏ các dòng có giá trị nằm ngoài giới hạn (outliers)
df_VIC_clean = df_VIC[~((df_VIC[['Open', 'Low', 'High', 'Close', 'Volume']] < lower_bound) | 
                    (df_VIC[['Open', 'Low', 'High', 'Close', 'Volume']] > upper_bound)).any(axis=1)].copy()

# Tính phần trăm thay đổi giữa giá mở cửa và giá đóng cửa
df_VIC_clean['Daily Change (%)'] = ((df_VIC_clean['Close'] - df_VIC_clean['Open']) / df_VIC_clean['Open']) * 100

# Tính phần trăm thay đổi giữa giá đóng cửa của các ngày liên tiếp
df_VIC_clean['Percent Change'] = df_VIC_clean['Close'].pct_change() * 100

# Lưu dữ liệu đã được làm sạch vào file CSV mới
df_VIC_clean.to_csv('clean_data_VIC_ADY.csv', index=False)

# In thông tin về số lượng dữ liệu trước và sau khi làm sạch
print(f"Số dòng dữ liệu ban đầu: {len(df_VIC)}")
print(f"Số dòng dữ liệu sau khi làm sạch: {len(df_VIC_clean)}\n")

print(df_VIC_clean)
````

3. **Tâm Lý Nhà Đầu Tư**: VNIndex phản ánh tâm lý và hành vi của nhà đầu tư trên thị trường. Các biến động lớn trong VNIndex có thể dẫn đến sự thay đổi trong nhu cầu mua bán cổ phiếu VIC, ảnh hưởng đến giá trị cổ phiếu.

### 2) Mô hình học sâu LSTM

LSTM (Long Short-Term Memory) là một dạng mạng nơ-ron hồi quy (RNN) cải tiến, giúp mô hình hóa tốt hơn các chuỗi thời gian dài và phức tạp. LSTM có khả năng ghi nhớ thông tin quan trọng trong khoảng thời gian dài hơn, giúp cải thiện độ chính xác khi dự báo giá cổ phiếu.

````python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

df = pd.read_csv('clean_data_VIC_ADY.csv')
df['tradeDate'] = pd.to_datetime(df['tradeDate'])
df.set_index('tradeDate', inplace=True)
df = df.asfreq('B')
df = df.ffill()
df = df[['Close']]
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)

def create_sequences(data, seq_length):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i:i+seq_length])
        y.append(data[i+seq_length])
    return np.array(X), np.array(y)

seq_length = 30
X, y = create_sequences(scaled_data, seq_length)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

print("Shape của X_train:", X_train.shape)  
print("Shape của y_train:", y_train.shape)
````

````python
import torch
import torch.nn as nn

class LSTMPredictor(nn.Module):
    def __init__(self, input_size=1, hidden_size=50, num_layers=2, output_size=1):
        super(LSTMPredictor, self).__init__()
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_size, output_size)

    def forward(self, x, return_hidden=False):  # Thêm option return_hidden
        lstm_out, _ = self.lstm(x)
        if return_hidden:
            return lstm_out  # Trả về hidden state
        return self.fc(lstm_out[:, -1, :])
    
model = LSTMPredictor()
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

# Kiểm tra thiết bị
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)  # Chuyển mô hình sang GPU nếu có

X_train_tensor_lstm = torch.tensor(X_train, dtype=torch.float32).to(device)
y_train_tensor_lstm = torch.tensor(y_train, dtype=torch.float32).to(device)
X_test_tensor_lstm = torch.tensor(X_test, dtype=torch.float32).to(device)
y_test_tensor_lstm = torch.tensor(y_test, dtype=torch.float32).to(device)
````

````python
# Chuyển dữ liệu vào DataLoader
from torch.utils.data import DataLoader, TensorDataset

num_epochs = 100  # Tăng epochs để hội tụ tốt hơn
batch_size = 32

train_dataset = TensorDataset(X_train_tensor_lstm, y_train_tensor_lstm)
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)

# Train loop
loss_history = []  # Lưu loss để vẽ biểu đồ

for epoch in range(num_epochs):
    model.train()  # Chế độ huấn luyện
    total_loss = 0
    
    for X_batch, y_batch in train_loader:
        X_batch, y_batch = X_batch.to(device), y_batch.to(device)

        optimizer.zero_grad()
        y_pred = model(X_batch)
        loss = criterion(y_pred, y_batch.view(-1, 1))
        
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    avg_loss = total_loss / len(train_loader)
    loss_history.append(avg_loss)

    if (epoch + 1) % 10 == 0:
        print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {avg_loss:.6f}")

torch.save(model.state_dict(), 'lstm_model.pth')
print("Mô hình đã được lưu.")
````

````python
model.eval()  

# Không tính gradient khi dự đoán
with torch.no_grad():
    y_test_pred = model(X_test_tensor_lstm.to(device))

# Chuyển kết quả về CPU và NumPy
y_test_pred = y_test_pred.cpu().numpy()
y_test_real = y_test_tensor_lstm.cpu().numpy()  # Chuyển tensor về NumPy trước

# Đảm bảo dữ liệu có shape đúng khi sử dụng inverse_transform
y_test_pred = scaler.inverse_transform(y_test_pred.reshape(-1, 1))  
y_test_real = scaler.inverse_transform(y_test_real.reshape(-1, 1))  

lstm_rmse = np.sqrt(mean_squared_error(y_test_real, y_test_pred))  
lstm_mae = mean_absolute_error(y_test_real, y_test_pred)
````

### 3) Mô hình học sâu Time Series Transformer

Transformer đã chứng minh hiệu quả trong xử lý chuỗi thời gian nhờ khả năng mô hình hóa quan hệ dài hạn giữa các điểm dữ liệu. Time Series Transformer (TST) là một mô hình dự báo chuỗi thời gian dựa trên Transformer Encoder, giúp cải thiện khả năng dự đoán giá cổ phiếu bằng cách học được các mẫu phức tạp trong dữ liệu lịch sử.

````python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import torch

df = pd.read_csv('clean_data_VIC_ADY.csv')
data = df[['Open', 'Low', 'High', 'Close', 'Volume', 'Daily Change (%)', 'Percent Change']]

scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)

# Tạo dataset
def create_dataset(data, time_step=60):
    X, y = [], []
    for i in range(len(data) - time_step - 1):
        X.append(data[i:(i + time_step)])
        y.append(data[i + time_step, 3])  # Dự đoán cột 'Close' (cột thứ 3)
    return np.array(X), np.array(y)

time_step = 60
X, y = create_dataset(scaled_data, time_step)

# Chia dữ liệu thành tập huấn luyện và tập kiểm tra
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# Chuyển dữ liệu thành tensor
X_train = torch.tensor(X_train, dtype=torch.float32)
y_train = torch.tensor(y_train, dtype=torch.float32)
X_test = torch.tensor(X_test, dtype=torch.float32)
y_test = torch.tensor(y_test, dtype=torch.float32)

nan_mask_X = torch.isnan(X_train).any(dim=(1, 2))  # Kiểm tra NaN ở bất kỳ chiều nào trong X_train
nan_mask_y = torch.isnan(y_train)  # Kiểm tra NaN trong y_train

# Tạo mặt nạ loại bỏ các chỉ số có NaN
nan_mask = nan_mask_X | nan_mask_y  # Chỉ cần bỏ đi nếu có NaN ở bất kỳ X_train hoặc y_train

# Áp dụng mặt nạ để loại bỏ các chỉ số có NaN
X_train_cleaned = X_train[~nan_mask]
y_train_cleaned = y_train[~nan_mask]

nan_mask_X_test = torch.isnan(X_test).any(dim=(1, 2))  # Kiểm tra NaN ở bất kỳ chiều nào trong X_test
nan_mask_y_test = torch.isnan(y_test)  # Kiểm tra NaN trong y_test

# Tạo mặt nạ loại bỏ các chỉ số có NaN
nan_mask_test = nan_mask_X_test | nan_mask_y_test  # Chỉ cần bỏ đi nếu có NaN ở bất kỳ X_test hoặc y_test

# Áp dụng mặt nạ để loại bỏ các chỉ số có NaN
X_test_cleaned = X_test[~nan_mask_test]
y_test_cleaned = y_test[~nan_mask_test]
````

````python
import torch.nn as nn

class TimeSeriesTransformer(nn.Module):
    def __init__(self, input_dim, time_step, d_model=64, n_heads=4, num_layers=2, output_dim=1):
        super(TimeSeriesTransformer, self).__init__()

        self.d_model = d_model

        # Lớp input projection
        self.input_projection = nn.Linear(input_dim, d_model)

        # Định nghĩa Encoder (Transformer)
        self.encoder = nn.TransformerEncoder(
            nn.TransformerEncoderLayer(d_model=d_model, nhead=n_heads),
            num_layers=num_layers
        )

        # Lớp output
        self.fc_out = nn.Linear(d_model, output_dim)

    def forward(self, x):
        # x có shape (batch_size, time_steps, input_dim)
        x = self.input_projection(x)  # Chuyển đổi kích thước về d_model
        x = x.permute(1, 0, 2)  # Chuyển x về (time_steps, batch_size, d_model)

        # Chạy qua Transformer Encoder
        x = self.encoder(x)

        # Lấy output cuối cùng
        x = x[-1, :, :]  # Chỉ lấy output của time_step cuối cùng
        x = self.fc_out(x)  # Dự đoán giá trị Close
        return x

# Xây dựng mô hình với các tham số tùy chỉnh
input_dim = X_train.shape[2]  # Số lượng đặc trưng
model = TimeSeriesTransformer(input_dim=input_dim, time_step=time_step, d_model=32, num_layers=1)

# Tạo optimizer và loss function
optimizer = torch.optim.Adam(model.parameters(), lr=1e-5)
criterion = nn.MSELoss()
````

````python
num_epochs = 200

# Huấn luyện mô hình
for epoch in range(num_epochs):
    model.train()

    # Đưa dữ liệu vào mô hình
    optimizer.zero_grad()
    output = model(X_train_cleaned)

    # Tính loss
    loss = criterion(output.squeeze(), y_train_cleaned)

    # Lan truyền ngược và tối ưu hóa
    loss.backward()
    torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)  # Gradient clipping
    optimizer.step()

    # In kết quả mỗi epoch
    if (epoch+1) % 10 == 0:
        print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}')
````

````python
model.eval()
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
X_test_cleaned = X_test_cleaned.to(device)
model.to(device)

# Không tính gradient khi dự đoán
with torch.no_grad():
    y_test_pred = model(X_test_cleaned)

# Chuyển tensor về NumPy trước khi sử dụng inverse_transform
y_test_pred = y_test_pred.cpu().numpy()  # Chuyển tensor về NumPy
y_test_real = y_test_cleaned.cpu().numpy()  # Chuyển tensor về NumPy

# Tạo mảng có 7 đặc trưng (bởi vì dữ liệu ban đầu có 7 cột)
y_test_pred_full = np.zeros((len(y_test_pred), data.shape[1]))  # Mảng 7 cột
y_test_real_full = np.zeros((len(y_test_real), data.shape[1]))  # Mảng 7 cột

# Chỉ gán giá trị của cột 'Close' (cột thứ 3) cho dự đoán và giá trị thực tế
y_test_pred_full[:, 3] = y_test_pred.reshape(-1)
y_test_real_full[:, 3] = y_test_real.reshape(-1)

# Sử dụng scaler.inverse_transform để phục hồi dữ liệu về giá trị gốc
# Dùng scaler đã được fit từ tập huấn luyện
y_test_pred_full = scaler.inverse_transform(y_test_pred_full)
y_test_real_full = scaler.inverse_transform(y_test_real_full)

# Chọn cột 'Close' (cột thứ 3) sau khi phục hồi giá trị
y_test_pred = y_test_pred_full[:, 3]
y_test_real = y_test_real_full[:, 3]

tst_rmse = np.sqrt(mean_squared_error(y_test_real, y_test_pred))  
tst_mae = mean_absolute_error(y_test_real, y_test_pred)
````

### 3) Phương pháp ICFTS

````python
def generate_counterfactual(instance, feature_idx, delta):
    """
    Tạo giải thích đối chứng bằng cách thay đổi một đặc trưng của instance.
    feature_idx: Vị trí đặc trưng cần thay đổi.
    delta: Giá trị thay đổi.
    """
    counterfactual_instance = instance.copy()
    counterfactual_instance[feature_idx] += delta
    return counterfactual_instance

def compute_change_percentage(original, counterfactual):
    """Tính phần trăm thay đổi của dự báo"""
    return ((counterfactual - original) / original) * 100
# Chọn một điểm dữ liệu cần giải thích
instance_cf = X_test[10].copy()

# Tạo 3 phiên bản đối chứng bằng cách thay đổi một đặc trưng
cf1 = generate_counterfactual(instance_cf, feature_idx=5, delta=2.0)
cf2 = generate_counterfactual(instance_cf, feature_idx=10, delta=-1.5)
cf3 = generate_counterfactual(instance_cf, feature_idx=15, delta=3.0)

# Dự đoán với các mẫu đối chứng
pred_original = predict_lstm(instance_cf.reshape(1, -1))
pred_cf1 = predict_lstm(cf1.reshape(1, -1))
pred_cf2 = predict_lstm(cf2.reshape(1, -1))
pred_cf3 = predict_lstm(cf3.reshape(1, -1))

# Tính phần trăm thay đổi
change_cf1 = compute_change_percentage(pred_original, pred_cf1)
change_cf2 = compute_change_percentage(pred_original, pred_cf2)
change_cf3 = compute_change_percentage(pred_original, pred_cf3)

# Hiển thị kết quả
print(f"Dự báo ban đầu: {pred_original}")
print(f"Dự báo đối chứng 1 (T_5 +2.0): {pred_cf1} ({change_cf1.item():.4f}%)")
print(f"Dự báo đối chứng 2 (T_10 -1.5): {pred_cf2} ({change_cf2.item():.4f}%)")
print(f"Dự báo đối chứng 3 (T_15 +3.0): {pred_cf3} ({change_cf3.item():.4f}%)")
````

````python
import matplotlib.pyplot as plt
import numpy as np

import matplotlib.pyplot as plt
import numpy as np

def plot_counterfactuals(original, counterfactuals, changes, labels, ax=None):
    """
    Vẽ biểu đồ so sánh dự báo gốc và các phiên bản đối chứng.
    
    Parameters:
    - original: Giá trị dự báo gốc (scalar).
    - counterfactuals: Danh sách các giá trị dự báo đối chứng.
    - changes: Danh sách phần trăm thay đổi tương ứng.
    - labels: Danh sách nhãn cho các đối chứng.
    - ax: Trục matplotlib để vẽ biểu đồ (tùy chọn).
    """
    # Nếu không có ax được truyền vào, tạo một figure và ax mới
    if ax is None:
        fig, ax = plt.subplots(figsize=(10, 5))
    
    # Tạo dữ liệu để vẽ
    x = np.arange(len(counterfactuals) + 1)  # Thêm một điểm cho dự báo gốc
    y = [original] + counterfactuals

    # Vẽ biểu đồ trên trục ax
    ax.plot(x, y, marker='o', linestyle='-', color='b', label="Dự báo")
    ax.set_xticks(x)
    ax.set_xticklabels(["Gốc"] + labels)
    
    # Hiển thị phần trăm thay đổi trên biểu đồ
    for i in range(1, len(y)):
        ax.text(x[i], y[i], f"{changes[i-1]:.2f}%", fontsize=12, color='r')

    # Thiết lập nhãn và tiêu đề
    ax.set_xlabel("Đối chứng")
    ax.set_ylabel("Giá trị dự báo")
    ax.set_title("So sánh dự báo giữa mẫu gốc và đối chứng (ICFTS)")
    ax.legend()
    ax.grid()

# Gọi hàm vẽ
plot_counterfactuals(
    pred_original.item(), 
    [pred_cf1.item(), pred_cf2.item(), pred_cf3.item()], 
    [change_cf1.item(), change_cf2.item(), change_cf3.item()],
    ["T_5+2.0", "T_10-1.5", "T_15+3.0"],ax=None
)
plt.show()
````

### 4) Phương pháp DAVOTS

````python
import torch
import torch.nn.functional as F

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = LSTMPredictor().to(device)
model.load_state_dict(torch.load('lstm_model.pth', weights_only=True)) 
model.eval()
def davots_attention(model, instance):
    """
    Tính Attention Score (DAVOTS) cho một instance cụ thể.
    """
    with torch.no_grad():
        lstm_out = model(instance.unsqueeze(0), return_hidden=True)  # Lấy toàn bộ hidden state
        pred = model(instance.unsqueeze(0), return_hidden=False)

        # Tính trọng số chú ý bằng dot-product
        attn_vector = model.fc.weight.T
        attn_scores = torch.bmm(lstm_out, attn_vector.unsqueeze(0))
        attn_scores = F.softmax(attn_scores, dim=1)  # Chuẩn hóa bằng Softmax

    return attn_scores.squeeze(0).cpu().numpy()  # Chuyển về numpy để trực quan hóa

# Chọn một instance từ tập test
instance_test = torch.tensor(X_test[100], dtype=torch.float32).to(device)

# Tính trọng số DAVOTS
attn_weights = davots_attention(model, instance_test)

# Vẽ biểu đồ trực quan hóa trọng số Attention
plt.figure(figsize=(10, 5))
plt.plot(attn_weights, marker='o', linestyle='-', color='b', label="Trọng số DAVOTS")
plt.xlabel("Thời điểm trong chuỗi")
plt.ylabel("Trọng số Attention")
plt.title("Phân bố trọng số theo thời gian (DAVOTS)")
plt.legend()
plt.grid()
plt.show()
````

**Đánh giá:**
DAVOTS (Deep Attention-based Visualization for Time Series) là phương pháp giúp trực quan hóa mức độ quan trọng của các thời điểm trong chuỗi thời gian đối với dự đoán của mô hình LSTM.

- Phân tích kết quả
  -  Thời điểm đầu tiên có trọng số cao nhất: Đây là một đặc điểm thường thấy trong mô hình LSTM, khi timestep đầu tiên chứa nhiều thông tin quan trọng ảnh hưởng đến dự báo.
  -  Trọng số giảm mạnh sau thời điểm đầu tiên: Điều này cho thấy mô hình chủ yếu tập trung vào thông tin gần đây, nhưng vẫn duy trì một mức độ ảnh hưởng nhất định từ các timestep sau đó.
  -  Sự ổn định trong vùng giữa và cuối chuỗi: Trọng số trong các timestep sau không thay đổi nhiều, cho thấy mô hình có xu hướng sử dụng toàn bộ chuỗi nhưng tập trung mạnh nhất vào các giá trị gần thời điểm đầu.

- Kết luận: DAVOTS giúp trực quan hóa sự tập trung của mô hình theo thời gian, từ đó cung cấp những thông tin quan trọng để tối ưu hóa và điều chỉnh mô hình LSTM một cách hiệu quả hơn.

````python
import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import plotly.graph_objs as go
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import shap
from lime import lime_tabular
import matplotlib.pyplot as plt
from io import BytesIO
import base64
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
class LSTMPredictor(nn.Module):
    def __init__(self, input_size=1, hidden_size=50, num_layers=2, output_size=1):
        super(LSTMPredictor, self).__init__()
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        lstm_out, _ = self.lstm(x)
        return self.fc(lstm_out[:, -1, :])

def create_sequences(data, seq_length):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i:i+seq_length])
        y.append(data[i+seq_length])
    return np.array(X), np.array(y)

# Đọc và xử lý dữ liệu
seq_length = 30
df = pd.read_csv('clean_data_VIC_ADY.csv')
data = df[['Close']].values  # Giả định cột 'Close' là giá đóng cửa
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data)

# Tạo chuỗi dữ liệu
X, y = create_sequences(scaled_data, seq_length)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# Chuyển đổi dữ liệu thực tế và dự đoán về thang gốc
y_test_actual = scaler.inverse_transform(y_test.reshape(-1, 1))

# Huấn luyện hoặc tải mô hình LSTM
model = LSTMPredictor().to(device)
model.load_state_dict(torch.load('lstm_model.pth', weights_only=True)) 
model.eval()

# Dự đoán trên tập kiểm tra
X_test_tensor = torch.tensor(X_test, dtype=torch.float32).to(device)
with torch.no_grad():
    y_pred_scaled = model(X_test_tensor).cpu().numpy()
y_pred = scaler.inverse_transform(y_pred_scaled)

# Tính giá trị SHAP
background = X_train[:100].reshape(-1, seq_length)
explainer = shap.KernelExplainer(
    lambda x: model(torch.tensor(x.reshape(-1, seq_length, 1), dtype=torch.float32).to(device)).detach().cpu().numpy(),
    background
)

X_test_2d = X_test.reshape(-1, seq_length)  # Biến đổi X_test thành 2D
computed_shap_values = {}

# Khởi tạo ứng dụng Dash
app = dash.Dash(__name__)

# Hàm tạo hình ảnh từ matplotlib để nhúng vào Dash
def matplotlib_to_base64(fig):
    buf = BytesIO()
    fig.savefig(buf, format="png", bbox_inches="tight")
    buf.seek(0)
    img_str = base64.b64encode(buf.getvalue()).decode("utf-8")
    buf.close()
    return f"data:image/png;base64,{img_str}"

# Hàm dự đoán LSTM
def predict_lstm(X):
    X = X.reshape(-1, seq_length, 1)
    X_tensor = torch.tensor(X, dtype=torch.float32).clone().detach().to(device)
    with torch.no_grad():
        predictions = model(X_tensor).cpu().numpy()
    return scaler.inverse_transform(predictions.reshape(-1, 1))

# Hàm phụ trợ cho ICFTS
def generate_counterfactual(instance, feature_idx, delta):
    cf_instance = instance.copy()
    cf_instance[feature_idx] += delta
    return cf_instance

def compute_change_percentage(original, counterfactual):
    return ((counterfactual - original) / original) * 100

def plot_counterfactuals(original, counterfactuals, changes, labels, ax):
    if ax is None:
        fig, ax = plt.subplots(figsize=(10, 5))
    x = np.arange(len(counterfactuals) + 1)
    y = [original] + counterfactuals
    ax.plot(x, y, marker='o', linestyle='-', color='b', label="Dự báo")
    ax.set_xticks(x)
    ax.set_xticklabels(["Gốc"] + labels[:3])
    for i in range(1, len(y)):
        ax.text(x[i], y[i], f"{changes[i-1]:.2f}%", fontsize=12, color='r')
    ax.set_xlabel("Đối chứng")
    ax.set_ylabel("Giá trị dự báo")
    ax.set_title("So sánh dự báo giữa mẫu gốc và đối chứng (ICFTS)")
    ax.legend()
    ax.grid()

# Hàm tính trọng số attention cho DAVOTS (giả lập)
def davots_attention(model, instance):
    instance = instance.reshape(1, seq_length, 1)
    with torch.no_grad():
        lstm_out, (hn, cn) = model.lstm(torch.tensor(instance, dtype=torch.float32).to(device))
    
    # Tính attention weights từ output của LSTM
    attention_weights = torch.softmax(lstm_out[:, -1, :].squeeze(), dim=0)
    return attention_weights.cpu().numpy()

# Chuẩn bị dữ liệu thời gian cho biểu đồ chính
time_index = pd.date_range(start='2024-01-01', periods=len(y_test_actual), freq='B')

# Layout của dashboard
app.layout = html.Div([
    html.H1("Dashboard Tương Tác Dự Đoán Giá Cổ Phiếu", style={'textAlign': 'center'}),
    dcc.Graph(id='main-graph'),
    
    html.Div([
        html.Div([
            html.H3("SHAP Explanation", className='text-center'),
            html.Img(id='shap-plot', style={'width': '100%'})
        ], className="six columns card"),
        
        html.Div([
            html.H3("LIME Explanation", className='text-center'),
            html.Img(id='lime-plot', style={'width': '100%'})
        ], className="six columns card"),
    ], className="row", style={'margin': '10px'}),
    
    html.Div([
        html.Div([
            html.H3("ICFTS Counterfactuals", className='text-center', style={'font-size': '16px'}),
            html.Img(id='icfts-plot', style={'width': '100%'})
        ], className="three columns card"),

        html.Div([
            html.H3("DAVOTS Attention", className='text-center', style={'font-size': '16px'}),
            html.Img(id='davots-plot', style={'width': '100%'})
        ], className="three columns card"),
], className="row", style={'margin': '10px', 'display': 'flex', 'justify-content': 'center'})

], style={'padding': '20px'})

# Callback để cập nhật biểu đồ chính và XAI khi nhấp vào điểm
@app.callback(
    [Output('main-graph', 'figure'),
     Output('shap-plot', 'src'),
     Output('lime-plot', 'src'),
     Output('icfts-plot', 'src'),
     Output('davots-plot', 'src')],
    [Input('main-graph', 'clickData')]
)
def update_graphs(clickData):
    fig_main = go.Figure()
    fig_main.add_trace(go.Scatter(x=time_index, y=y_test_actual.flatten(), mode='lines', name='Thực tế'))
    fig_main.add_trace(go.Scatter(x=time_index, y=y_pred.flatten(), mode='lines', name='Dự đoán'))
    fig_main.update_layout(title='Giá Thực Tế vs Dự Đoán', xaxis_title='Thời gian', yaxis_title='Giá đóng cửa')

    selected_idx = 0 if clickData is None else clickData['points'][0]['pointIndex']

    # 1. SHAP
    if selected_idx not in computed_shap_values:
        shap_val = explainer.shap_values(X_test_2d[selected_idx:selected_idx+1].copy(), nsamples=500)[0]
        computed_shap_values[selected_idx] = shap_val
    shap_vals_single = computed_shap_values[selected_idx].reshape(1, -1)
    
    plt.figure(figsize=(6, 4))
    shap.force_plot(
        explainer.expected_value, 
        shap_vals_single[0],
        features=X_test_2d[selected_idx],
        feature_names=[f"T-{i}" for i in range(seq_length, 0, -1)],
        matplotlib=True,
        show=False
    )
    plt.tight_layout()
    shap_img = matplotlib_to_base64(plt.gcf())
    plt.close()


    # 2. LIME
    explainer_lime = lime_tabular.LimeTabularExplainer(
        training_data=X_train.reshape(-1, seq_length),
        feature_names=[f"T_{i}" for i in range(seq_length)],
        mode="regression",
        kernel_width=2,
        discretize_continuous=False
    )
    instance = X_test[selected_idx].reshape(1, seq_length)
    explanation = explainer_lime.explain_instance(instance.flatten(), predict_lstm)
    fig_lime = explanation.as_pyplot_figure()
    lime_img = matplotlib_to_base64(fig_lime)
    plt.close(fig_lime)

    # 3. ICFTS
    instance_cf = X_test[selected_idx].copy()
    shap_values_flat = shap_vals_single.flatten()
    top_features = np.argsort(np.abs(shap_values_flat))[-3:]

    # Tạo counterfactuals dựa trên các feature quan trọng nhất
    cf_instances = []
    deltas = [0.02, -0.02, 0.05]
    labels = []
    for i, feat_idx in enumerate(top_features):
        delta = deltas[i % len(deltas)]
        cf = generate_counterfactual(instance_cf, feat_idx, delta)
        cf_instances.append(cf)
        labels.append(f"T-{feat_idx} {delta:+.2f}")

    # Phần còn lại giữ nguyên
    pred_original = predict_lstm(instance_cf.reshape(1, -1))
    pred_cf1 = predict_lstm(cf_instances[0].reshape(1, -1))  # Thay cf1 bằng cf_instances[0]
    pred_cf2 = predict_lstm(cf_instances[1].reshape(1, -1))  # Thay cf2 bằng cf_instances[1]
    pred_cf3 = predict_lstm(cf_instances[2].reshape(1, -1))  # Thay cf3 bằng cf_instances[2]
    change_cf1 = compute_change_percentage(pred_original, pred_cf1)
    change_cf2 = compute_change_percentage(pred_original, pred_cf2)
    change_cf3 = compute_change_percentage(pred_original, pred_cf3)

    pred_original_scalar = pred_original[0][0]
    pred_cf1_scalar = pred_cf1[0][0]
    pred_cf2_scalar = pred_cf2[0][0]
    pred_cf3_scalar = pred_cf3[0][0]
    change_cf1_scalar = change_cf1[0][0]
    change_cf2_scalar = change_cf2[0][0]
    change_cf3_scalar = change_cf3[0][0]

    fig_icfts, ax = plt.subplots(figsize=(6, 4))
    plot_counterfactuals(pred_original_scalar, 
                         [pred_cf1_scalar, pred_cf2_scalar, pred_cf3_scalar], 
                         [change_cf1_scalar, change_cf2_scalar, change_cf3_scalar], 
                         labels, ax=ax)
    icfts_img = matplotlib_to_base64(fig_icfts)
    plt.close(fig_icfts)

    # 4. DAVOTS
    instance_test = torch.tensor(X_test[selected_idx], dtype=torch.float32).to(device)
    attn_weights = davots_attention(model, instance_test)
    fig_davots, ax = plt.subplots(figsize=(6, 4))
    ax.plot(attn_weights, marker='o', linestyle='-', color='b')
    ax.set_title('Trọng số DAVOTS')
    ax.set_xlabel('Thời điểm trong chuỗi')
    ax.set_ylabel('Trọng số Attention')
    davots_img = matplotlib_to_base64(fig_davots)
    plt.close(fig_davots)

    return fig_main, shap_img, lime_img, icfts_img, davots_img

# Chạy ứng dụng
if __name__ == '__main__':
    app.run_server(debug=True)
````

### I) Phân tích tài chính

#### Dựa trên kết quả phân tích mô hình dự báo giá cổ phiếu, có thể rút ra một số kết luận quan trọng về các yếu tố ảnh hưởng đến thị trường tài chính:

1) Tầm quan trọng của dữ liệu gần nhất

- Kết quả từ SHAP và DAVOTS cho thấy các timestep gần nhất (T-1, T-2, T-3) có ảnh hưởng lớn đến dự đoán giá cổ phiếu, phản ánh tính chất ngắn hạn của thị trường tài chính.
- Điều này phù hợp với quan sát thực tế: giá cổ phiếu thường bị chi phối mạnh bởi các biến động gần đây hơn là dữ liệu xa trong quá khứ.

2) Ảnh hưởng của dữ liệu lịch sử xa hơn

- Phân tích bằng LIME và ICFTS cho thấy một số timestep xa hơn (T-10, T-15, T-29) vẫn có tác động đáng kể đến dự báo, mặc dù nhỏ hơn so với các timestep gần.
- Điều này gợi ý rằng một số yếu tố trong quá khứ có thể mang lại thông tin quan trọng, chẳng hạn như chu kỳ kinh tế hoặc ảnh hưởng của các sự kiện lớn (ví dụ: báo cáo tài chính quý, chính sách tiền tệ).

3) Sự khác biệt trong cách mô hình xử lý thông tin

- SARIMA chủ yếu tập trung vào xu hướng và tính chu kỳ của dữ liệu, phù hợp với các thị trường có mô hình biến động rõ ràng.
- LSTM và TST tận dụng thông tin từ nhiều timestep khác nhau, nhưng LSTM có xu hướng dựa vào dữ liệu gần hơn, trong khi TST có thể khai thác thông tin từ quá khứ xa hơn thông qua cơ chế Attention.
- Điều này phản ánh hai chiến lược đầu tư khác nhau: giao dịch ngắn hạn dựa trên biến động gần và đầu tư dài hạn dựa trên xu hướng dài hạn.

4) Sự nhạy cảm của mô hình đối với dữ liệu đầu vào

- ICFTS cho thấy một số khoảng thời gian nhất định có ảnh hưởng lớn đến dự báo, điều này có thể liên quan đến các yếu tố kinh tế vĩ mô hoặc sự kiện thị trường quan trọng.
- Điều này gợi ý rằng việc lựa chọn dữ liệu đầu vào phù hợp và xử lý các yếu tố tác động bên ngoài có thể cải thiện hiệu suất dự báo của mô hình.
